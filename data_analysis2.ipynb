```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab

# Pour éviter d'avoir à préciser ces informations à chaque graphique
params = {'legend.fontsize': 'x-large',
          'figure.figsize': (15, 15),
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large'}
pylab.rcParams.update(params)
```

```python
data = []

def add_data(arr, i):
    df = pd.read_csv ('data/data_no_dh_1000_fine/generation'+str(i)+'.csv', skiprows=0, 
                          index_col=0, header=0)
    arr.append([i,
                df['h'].to_numpy(),
                df['v'].to_numpy(),
                df['fuel'].to_numpy()])

idx = [0, 1, 11, 21, 24, 28, 48]
for i in range(len(idx)):
    add_data(data, idx[i])
```

```python
for i in range(len(data)):
    if i % 2 == 0:
        plt.plot(data[i][1], data[i][2], label="Génération " + str(data[i][0]), linestyle="dashed", linewidth=1)
    else:
        plt.plot(data[i][1], data[i][2], label="Génération " + str(data[i][0]), linestyle="solid", linewidth=2)
       
plt.xlabel(r'$h$ [m]')
plt.ylabel(r'$v$ [m s$^{-1}$]')
plt.title(r"Convergence dans l'espace des phases")
plt.grid()
plt.legend(loc='upper right')
plt.show()
```

```python
for i in range(len(data)):
    if i % 2 == 0:
        plt.plot(data[i][1], data[i][3], label="Génération " + str(data[i][0]), linestyle="dashed", linewidth=1)
    else:
        plt.plot(data[i][1], data[i][3], label="Génération " + str(data[i][0]), linestyle="solid", linewidth=2)

plt.xlabel(r'$h$ [m]')
plt.ylabel(r'Carburant restant [kg]')
plt.title(r"Carburant utilisé en fonction de l'altitude")
plt.grid()
plt.legend(loc='upper right')
plt.show()
```
